```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load csv file
student_data = pd.read_csv('E:/DQLab/Dataset/Datacamp/student-alcohol-consumption.csv')

# Display top 5 rows
print(student_data.head())
```

```python
# Display dataset information
print(student_data.info(memory_usage=False))
```

```python
# Display summary statistics 
print(student_data.describe())
```

```python
# Display null value in each column
print(student_data.isnull().any())
```

```python
sns.heatmap(student_data.corr(), annot = True, fmt = ".2f", cbar = True)
plt.gcf().set_size_inches(15, 10)
plt.savefig('1_heatmap.png')
```

```python
sns.countplot(x='school', data=student_data, palette='Set2', hue='location')
plt.title("School Distribution", fontsize=20, pad=15)
plt.gcf().set_size_inches(8, 6)
plt.savefig('2_school dist.png')
plt.show()
```

```python
sex = student_data['sex'].value_counts()
label = ['Female', 'Male']
plt.pie(sex, labels = label, autopct='%1.2f%%')
plt.title("Gender Distribution", fontsize=20, pad=10)
plt.gcf().set_size_inches(5, 5)
plt.savefig('3_sex dist.png')
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="sex", y="Walc",
            data=student_data,
            palette="Set2",
            kind="bar",
            ci=None,
            order=["M", "F"])

plt.title("Weekend Alcohol Consumption by Gender", fontsize=20, pad=10)
plt.gcf().set_size_inches(5, 6)
plt.savefig('4_weekend alc sex.png')
# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="sex", y="Dalc",
            data=student_data,
            palette="Set2",
            kind="bar",
            ci=None,
            order=["M", "F"])

plt.title("Weekday Alcohol Consumption by Gender", fontsize=20, pad=10)
plt.gcf().set_size_inches(6, 5)
plt.savefig('5_weekday alc sex.png')
# Show plot
plt.show()
```

```python
sns.countplot(x='age', data=student_data, palette="Set2")
plt.title("Age Distribution", fontsize=20, pad=15)
plt.gcf().set_size_inches(9, 6)
plt.savefig('6_age dist.png')
plt.show()
```

```python
famsize = student_data['famsize'].value_counts()
label = ['More than 3', 'Less or equal 3']
plt.pie(famsize, labels = label, autopct='%1.2f%%')
plt.title("Family Size Distribution", fontsize=20, pad=10)
plt.gcf().set_size_inches(5, 5)
plt.savefig('7_famsize dist.png')
plt.show()
```

```python
pstatus = student_data['Pstatus'].value_counts()
label = ['Together', 'Apart']
plt.pie(pstatus, labels = label, autopct='%1.2f%%')
plt.title("Parent's status", fontsize=20, pad=10)
plt.gcf().set_size_inches(5, 5)
plt.savefig('8_pstatus dist.png')
plt.show()
```

```python
sns.countplot(x='study_time', 
              data=student_data, 
              palette="Set2", 
              order=["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"])
plt.title("Study Time Distribution", fontsize=20, pad=15)
plt.gcf().set_size_inches(8, 5)
plt.savefig('9_studytime dist.png')
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            palette="Set2",
            kind="bar",
            ci=None,
            order=["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"])

plt.title("Correlation Study Time and Final Grade", fontsize=20, pad=15)
plt.gcf().set_size_inches(8, 5)
plt.savefig('10_study G3.png')
# Show plot
plt.show()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x="study_time", y="G3", 
            data=student_data, 
            kind="box", 
            order=study_time_order)

plt.title("Correlation Study Time and Final Grade", fontsize=20, pad=15)
plt.gcf().set_size_inches(7, 6)
plt.savefig('11_study G3 boxplot.png')
# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="goout", y="freetime",
            data=student_data,
            kind="point",
            ci=None)
plt.title("Correlation Having a Free Time and Student Go Out", fontsize=20, pad=15, loc="center")
plt.gcf().set_size_inches(7, 6)
plt.savefig('12_freetime goout.png')
# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="goout", y="Walc",
            data=student_data,
            kind="point",
            ci=None)
plt.title("Correlation Weekend Alcohol Consumption and Student Go Out", fontsize=20, pad=15, loc="center")
plt.gcf().set_size_inches(7, 6)
plt.savefig('12_goout Walc.png')
# Show plot
plt.show()
```

```python
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data,
                hue="location",
                hue_order=["Rural", "Urban"])

plt.title("Correlation Absences and Final Grade", fontsize=20, pad=15)
plt.gcf().set_size_inches(7, 5)
plt.savefig('13_abs G3 loc.png')
# Show plot
plt.show()
```

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x="G1", y="G3", 
            data=student_data, 
            kind="scatter")

plt.title("Correlation 1st Grade and Final Grade", fontsize=20, pad=15)
plt.savefig('14_G1 G3.png')
# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup", 
            col_order=["yes", "no"])

plt.title("Correlation 1st Grade Final Grade by School Support", fontsize=20, pad=15)
plt.savefig('15_G1 G3 col.png')
# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])

plt.gcf().set_size_inches(13, 9)
plt.savefig('15_G1 G3 col row.png')
# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", 
            data=student_data, 
            kind="box", 
            sym="", 
            hue="location")

plt.title("Correlation Internet Access and Final Grade", fontsize=20, pad=15)
plt.savefig('16_int G3.png')
# Show plot
plt.show()
```

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box")

plt.title("Correlation Romantic and Final Grade", fontsize=20, pad=15)
plt.savefig('17_rom G3.png')
# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2)
plt.title("Correlation Family Relationship and Absences", fontsize=20, pad=15)
plt.gcf().set_size_inches(7,6)
plt.savefig('18_famrel abs.png') 
# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="G3",
            data=student_data,
            kind="point",
            capsize=0.2)

plt.title("Correlation Family Relationship and Final Grade", fontsize=20, pad=15)
plt.gcf().set_size_inches(7,6)
plt.savefig('19_famrel G3.png')
# Show plot
plt.show()
```

```python
# Create a point plot with subgroups
sns.catplot(x="romantic", y="absences", 
            data=student_data, 
            kind="point", hue="school")

plt.title("Correlation Romantic and Absences", fontsize=20, pad=15)
plt.gcf().set_size_inches(7,6)
# Show plot
plt.show()
```

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None)

plt.title("Correlation Romantic and Absences", fontsize=20, pad=15)
plt.savefig('20_rom abs.png')
# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator=median)

plt.title("Correlation Romantic and Absences", fontsize=20, pad=15)
plt.savefig('21_rom abs mean.png')
# Show plot
plt.show()
```
